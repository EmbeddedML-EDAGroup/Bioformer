```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np

N, C = 30, 64
a = np.empty((N, C))
for i in range(N):
    a[i, 0::2][::-1] = np.sin(i / (10000 ** (2*np.arange(1, C, 2)/C)))
    a[i, 1::2][::-1] = np.cos(i / (10000 ** (2*np.arange(1, C, 2)/C)))

fig, ax = plt.subplots(figsize=(7, 5))
im = ax.imshow(a, extent=[C,1,N,0], cmap='cividis')


ax.xaxis.set_major_locator(ticker.MultipleLocator(8))
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter("%dth"))
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
#ax.yaxis.set_major_formatter(ticker.FormatStrFormatter("%dth"))

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="2%", pad=0.2)
   
plt.colorbar(im, cax=cax)

ax.set_xlabel('Embedding dimension')
ax.set_ylabel('Token position')

fig.savefig('pos-embedding.png', format='png', dpi=300, bbox_inches='tight')
```
