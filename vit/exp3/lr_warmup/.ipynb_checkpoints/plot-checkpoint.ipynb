```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pickle import load
```

```python
configs, n_params, losses_accs_ = load(open('warmup.pickle', 'rb'))
model_config = configs[0]
losses_accs = losses_accs_[0]
losses_accs_df_w = pd.DataFrame(losses_accs)

configs, n_params, losses_accs_ = load(open('fixed1e-3.pickle', 'rb'))
model_config_ = configs[0]
losses_accs = losses_accs_[0]
losses_accs_df_f = pd.DataFrame(losses_accs)

n_params = n_params[0]
```

```python
model_config == model_config_
```

```python
fig, axes = plt.subplots(figsize=(16, 8), nrows=2, ncols=2)

print("Soggetto: 5")
print(model_config)
print("parametri:", f"{n_params / 1e3:.2f}k")

ax = axes[0, :]
ax[0].plot(losses_accs_df_w['train_acc'], label='train_acc')
ax[0].plot(losses_accs_df_w['val_acc'], label='val_acc')
ax[0].plot(losses_accs_df_w['test_acc'], label='test_acc')
ax[0].grid(True)
ax[0].legend()
ax[0].set_title('Warmup')

ax[1].plot(losses_accs_df_f['train_acc'], label='train_acc')
ax[1].plot(losses_accs_df_f['val_acc'], label='val_acc')
ax[1].plot(losses_accs_df_f['test_acc'], label='test_acc')
ax[1].grid(True)
ax[1].legend()
ax[1].set_title('Fisso 1e-3 dimezzato ogni 50')
ax[0].set_ylim([.4, 1])
ax[1].set_ylim([.4, 1])

ax = axes[1, :]
ax[0].plot(losses_accs_df_w['train_loss'], label='train_loss')
ax[0].plot(losses_accs_df_w['val_loss'], label='val_loss')
ax[0].plot(losses_accs_df_w['test_loss'], label='test_loss')
ax[0].grid(True)
ax[0].legend()

ax[1].plot(losses_accs_df_f['train_loss'], label='train_loss')
ax[1].plot(losses_accs_df_f['val_loss'], label='val_loss')
ax[1].plot(losses_accs_df_f['test_loss'], label='test_loss')
ax[1].grid(True)
ax[1].legend()
ax[0].set_ylim([0, 2.75])
ax[1].set_ylim([0, 2.75])

None
```

```python
import torch
net = torch.nn.Linear(100,2)
optimizer = torch.optim.Adam(net.parameters())
scheduler = torch.optim.lr_scheduler.CyclicLR(optimizer, base_lr=1e-7, max_lr=1e-3, step_size_up=100, step_size_down=None, mode='triangular', cycle_momentum=False)
lrs = []
for i in range(200):
    lrs.append(scheduler.get_lr())
    scheduler.step()
```

```python
plt.plot(lrs)
```
