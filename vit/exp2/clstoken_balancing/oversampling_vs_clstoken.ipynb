```python
from pickle import load
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
configs, losses_accs_ = load(open('train_bootstrap_vs_clstoken.pickle', 'rb'))
```

```python
losses_accs_df = list(map(lambda x: pd.DataFrame(x), losses_accs_))
```

```python
for i in list(range(len(configs)))[::2]:
    plt.plot(losses_accs_df[i]['val_acc'], color="blue")
    plt.plot(losses_accs_df[i]['test_acc'], color="blue", label=f"class_token: {configs[i]['use_cls_token']}")
    plt.plot(losses_accs_df[i+1]['val_acc'], color="red")
    plt.plot(losses_accs_df[i+1]['test_acc'], color="red",  label=f"class_token: {configs[i+1]['use_cls_token']}")
    plt.grid()
    plt.legend()
    plt.title(configs[i])
    plt.ylim([.5, 1])
    plt.show()
```
