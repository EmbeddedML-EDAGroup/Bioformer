```python
from pickle import load
import matplotlib.pyplot as plt
```

```python
patch_sizes, losses_accs_ = load(open('losses_accs_1e-3_30epochs_.2drop_1each-epoch_1sess_okinit.pickle', 'rb'))

plt.plot(list(map(str, patch_sizes)), [losses_accs.val_acc.max() for losses_accs in losses_accs_], marker="o", label="Validation S1")
plt.plot(list(map(str, patch_sizes)), [losses_accs.test_acc.max() for losses_accs in losses_accs_], marker="o", label="Test S2")
plt.axhline(.85, color="C0")
plt.axhline(.65, color="C1")
plt.title("1 session subject 5")
plt.legend()
plt.grid()
plt.show()
```

```python
patch_sizes, losses_accs_ = load(open('losses_accs_1e-3_30epochs_.2drop_1each-epoch_5sess_okinit.pickle', 'rb'))

plt.plot(list(map(str, patch_sizes)), [losses_accs.val_acc.max() for losses_accs in losses_accs_], marker="o", label="Validation S1-5")
plt.plot(list(map(str, patch_sizes)), [losses_accs.test_acc.max() for losses_accs in losses_accs_], marker="o", label="Test S6")
plt.axhline(.835, color="C0")
plt.axhline(.834, color="C1")
plt.title("5 sessions subject 5")
plt.legend()
plt.grid()
plt.show()
```
