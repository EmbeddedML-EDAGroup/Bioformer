```python
from pickle import load

configs, n_params_, losses_accs_ = load(open('results_1623206515.pickle', 'rb'))

import matplotlib.pyplot as plt
import pandas as pd

losses_accs_df = list(map(lambda x: pd.DataFrame(x), losses_accs_))

for config, losses_accs in zip(configs, losses_accs_df):
    
    print("Addestrato su ", config['sessions'], "sessioni con i soggetti", config['subjects'])
    print(config['training_config']['lr_scheduler'])
    print(config['dropout'])
    print(config['emb_dropout'])
    #print(config)
    
    fig, ax = plt.subplots(figsize=(12,4), ncols=2)
    ax[0].plot(losses_accs['train_acc'])
    ax[0].plot(losses_accs['val_acc'])
    ax[0].plot(losses_accs['test_acc'])
    ax[0].set_ylim([.5, 1.])
    
    ax[1].plot(losses_accs['train_loss'])
    ax[1].plot(losses_accs['val_loss'])
    ax[1].plot(losses_accs['test_loss'])
    ax[1].set_ylim([0, 4])
    
    ax[0].grid()
    ax[1].grid()
    
    plt.show()
```
